```python
import hepmcanalysis
import ROOT
import rootnotes
```

```python
h = ROOT.TH1F('finalstatept','finalstatept',50,0,500)
```

```python
for e in hepmcanalysis.readhepmc('input.hepmc'):
    for p in e.fsParticles():
        h.Fill(p.momentum().perp())
```

```python
c = rootnotes.default_canvas()
```

```python
h.SetFillColor(38)
h.Draw()
```
